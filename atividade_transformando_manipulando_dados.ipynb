```python
import pandas as pd
import numpy as np
```

```python
url1 = 'https://raw.githubusercontent.com/Migueuu17/atividade_tranformando_dados_numericos/refs/heads/main/dados_locacao_imoveis.json'
url2 = 'https://raw.githubusercontent.com/Migueuu17/atividade_tranformando_dados_numericos/refs/heads/main/dados_vendas_clientes.json'
```

```python
dados_locacao_imoveis = pd.read_json(url1)
dados_vendas_clientes = pd.read_json(url2)
```

```python
dados_locacao_imoveis = pd.json_normalize(dados_locacao_imoveis['dados_locacao'])
```

```python
dados_vendas_clientes = pd.json_normalize(dados_vendas_clientes['dados_vendas'])
```

#Projeto desafio 1: vendas online

1. Remover os dados em listas dentro do DataFrame;
2. Verificar os tipos de dados;
3. Identificar colunas numéricas;
4. Transformar a coluna numérica para o tipo numérico.
5. Manipular os textos presentes na coluna Cliente para que seja obtido como resultado os nomes dos clientes em letras minúsculas, com a ausência de caracteres especiais ou números.
6. Transformar dados de datas e criar visualizações estratégicas

```python
dados_vendas_clientes.head()
```

```python
colunas = list(dados_vendas_clientes.columns)
colunas
```

```python
dados_vendas_clientes = dados_vendas_clientes.explode(colunas[1:])
```

```python
dados_vendas_clientes.reset_index(inplace=True, drop=True)
```

```python
dados_vendas_clientes.head()
```

```python
dados_vendas_clientes.info()
```

```python
dados_vendas_clientes['Valor da compra'] = dados_vendas_clientes['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','.').strip())
```

```python
dados_vendas_clientes['Valor da compra'] = dados_vendas_clientes['Valor da compra'].astype(np.float64)
```

```python
dados_vendas_clientes.info()
```

```python
dados_vendas_clientes.head()
```

```python
dados_vendas_clientes['Cliente'] = dados_vendas_clientes['Cliente'].str.replace('[0-9]', '', regex=True)
```

```python
dados_vendas_clientes['Cliente'] = dados_vendas_clientes['Cliente'].str.replace('@', '')
```

```python
dados_vendas_clientes['Cliente'] = dados_vendas_clientes['Cliente'].str.replace('_', ' ')
```

```python
dados_vendas_clientes['Cliente'] = dados_vendas_clientes['Cliente'].str.upper()
```

```python
dados_vendas_clientes
```

```python
dados_vendas_clientes['Data de venda'] = pd.to_datetime(dados_vendas_clientes['Data de venda'], format= '%d/%m/%Y')
```

```python
dados_vendas_clientes['Data de venda'].dt.strftime('%d/%m')
```

```python
subset = dados_vendas_clientes.groupby(dados_vendas_clientes['Data de venda'].dt.strftime('%d/%m'))['Valor da compra'].sum()
subset
```

#Projeto desafio 2: administração de condomínios

1. Remover os dados em listas dentro do DataFrame;
2. Verificar os tipos de dados;
3. Identificar colunas numéricas;
4. Transformar a coluna numérica para o tipo numérico.
5. Manipular os textos na coluna apartamento para remover o texto (blocoAP) do DataFrame.
6. Transformar dados de datas e criar visualizações estratégicas

```python
dados_locacao_imoveis
```

```python
colunas = list(dados_locacao_imoveis.columns)
colunas
```

```python
dados_locacao_imoveis = dados_locacao_imoveis.explode(colunas[1:])
```

```python
dados_locacao_imoveis.reset_index(inplace=True,drop=True)
```

```python
dados_locacao_imoveis
```

```python
dados_locacao_imoveis.info()
```

```python
dados_locacao_imoveis['valor_aluguel'] = dados_locacao_imoveis['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',','.').replace('reais', '').strip())
```

```python
dados_locacao_imoveis['valor_aluguel'] = dados_locacao_imoveis['valor_aluguel'].astype(np.float64)
```

```python
dados_locacao_imoveis.info()
```

```python
dados_locacao_imoveis['apartamento'] = dados_locacao_imoveis['apartamento'].str.replace(' (blocoAP)', '')
```

```python
dados_locacao_imoveis.head()
```

```python
dados_locacao_imoveis['datas_combinadas_pagamento'] = pd.to_datetime(dados_locacao_imoveis['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados_locacao_imoveis['datas_de_pagamento'] = pd.to_datetime(dados_locacao_imoveis['datas_de_pagamento'], format='%d/%m/%Y')
```

```python
dados_locacao_imoveis['status_pagamento'] = dados_locacao_imoveis.apply(
    lambda row: 'Atrasado' if row['datas_de_pagamento'] > row['datas_combinadas_pagamento'] else 'No Prazo',
    axis=1
)
```
